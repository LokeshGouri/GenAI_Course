```python
def checkCholestoral(cholestoralValues, threshold):
  output = []
  if(isinstance(cholestoralValues, pd.Series)):
    for value in cholestoralValues:
      if(value > threshold):
        output.append(True)
      else:
        output.append(False)
  return output

print(checkCholestoral(df["cholestoral"], 400))

```
